```octave
pkg load image
image=imread('file:///home/lenovo/Pictures/Screenshot%20from%202020-05-08%2016-49-25.png');
img=rgb2gray(image);
imshow(img)
```

```octave
edges=edge(img,'canny');
imshow(edges)
```

```octave
[acc,theta,rho]=hough(edges);
```

```octave

figure;imagesc(acc,'XData',theta,'YData',rho),title('points');
```

```octave
peaks=houghpeaks(acc,100,'Threshold',ceil(0.6*max(acc(:))),'NHoodSize',[5,5]);

plot(theta(peaks(:,2)),rho(peaks(:,1)),'rhs');
```

```octave
size(peaks)
```

```octave
lines_seg=houghlines(edges,theta,rho,peaks,'FillGap',50,'MinLength',100);
lines_seg
```

```octave
length(lines_seg)
```

```octave
figure;imshow(img);
hold on;
for k=1:length(lines_seg)
    endpoints=[lines_seg(k).point1;lines_seg(k).point2];
    plot(endpoints(:,1),endpoints(:,2),'linewidth',2,'color','green');
end
hold off;
```
